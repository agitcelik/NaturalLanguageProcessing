```python
# *** Preprocess the Features

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# -> TOKENIZATION with PUNCUTATION REMOVAL
tokenizedReviewsTextList = list()
tokenizer = RegexpTokenizer(r'\w+')

for review in X.values.astype('U'):
    tokenizedReviewsTextList.append(tokenizer.tokenize(review))

# -> STOPWORDS REMOVAL
stop_words = set(stopwords.words('english'))

currentFilteredTokenizedReviewsTextList = list()
filteredTokenizedReviewsTextList = list()

# Nested for -- because of the fact that filteredTokenizedReviewsTextList is array of arrays:
for review in tokenizedReviewsTextList:
    for tokenizedWord in review:
        tokenizedWord = tokenizedWord.lower()
        if tokenizedWord not in stop_words:
            currentFilteredTokenizedReviewsTextList.append(tokenizedWord)

    # Keep the data into filteredTokenizedReviewsTextList, for further purposes
    filteredTokenizedReviewsTextList.append(currentFilteredTokenizedReviewsTextList.copy())

    currentFilteredTokenizedReviewsTextList.clear()

# -> Porter Stemmer:
ps = PorterStemmer()

currentPStemmedFilteredTokenizedReviewsTextList = list()
pStemmedFilteredTokenizedReviewsTextList = list()

for review in filteredTokenizedReviewsTextList:
    for filteredTokenizedWord in review:
        filteredTokenizedWord = ps.stem(filteredTokenizedWord)
        currentPStemmedFilteredTokenizedReviewsTextList.append(filteredTokenizedWord)

    pStemmedFilteredTokenizedReviewsTextList.append(
        currentPStemmedFilteredTokenizedReviewsTextList.copy())

    currentPStemmedFilteredTokenizedReviewsTextList.clear()
```

```python
# -> Obtain the requested text_data for creation of the Bag-of-Words
text_data = []
review_sentences = ""
for review in pStemmedFilteredTokenizedReviewsTextList:
    for word in review:
        review_sentences = review_sentences +word +" "
    text_data.append(review_sentences)
    review_sentences = ""
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create the bag of words feature matrix
count = CountVectorizer()
bag_of_words = count.fit_transform(text_data)

# Show feature matrix
bag_of_words.toarray()
```

```python
# Get feature names
feature_names = count.get_feature_names()

# View feature names
feature_names
```

```python
# Create data frame
X = pd.DataFrame(bag_of_words.toarray(), columns=feature_names)
X
```

```python
# *** Preparation for ML
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
# default split is 75% for training and 25% for testing
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## <font color=red> WordNET </font>

```python
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet as wn
import random

review_data = df['reviews_text'] #Seperate the comments
review_data
```

```python
review_data_list = review_data.tolist()
```

```python
def penn_to_wn(tag):
    """ Convert between a Penn Treebank tag to a simplified Wordnet tag """
    if tag.startswith('N'):
        return 'n'
 
    if tag.startswith('V'):
        return 'v'
 
    if tag.startswith('J'):
        return 'a'
 
    if tag.startswith('R'):
        return 'r'
 
    return None
```

```python
def tagged_to_synset(word, tag):
    wn_tag = penn_to_wn(tag)
    if wn_tag is None:
        return None
 
    try:
        return wn.synsets(word, wn_tag)[0]
    except:
        return None
```

```python
def sentence_similarity(sentence1, sentence2):
    """ compute the sentence similarity using Wordnet """
    # Tokenize and tag
    
    print("SENTENCE-1= ", sentence1)
    print()
    
    # Get the synsets for the tagged words
    sentence1 = nltk.pos_tag(word_tokenize(sentence1))
    print("TAGS OF SENTENCE-1= ", sentence1)
    print()
    
    synsets1 = [tagged_to_synset(*tagged_word) for tagged_word in sentence1]
    print("SYNSETS OF SENTENCE-1= ", synsets1)
    
    
    print("___")
    print()
    
    print("SENTENCE-2= ", sentence2)
    print()
    
    # Get the synsets for the tagged words
    sentence2 = nltk.pos_tag(word_tokenize(sentence2))
    print("TAGS OF SENTENCE-2= ", sentence2)
    print()
    
    synsets2 = [tagged_to_synset(*tagged_word) for tagged_word in sentence2]
    print("SYNSETS OF SENTENCE-2= ", synsets2)
    
    # Filter out the Nones
    synsets1 = [ss for ss in synsets1 if ss]
    synsets2 = [ss for ss in synsets2 if ss]
 
    score, count = 0.0, 0
    best_score = 0
    arr_simi_score = []
    for syn1 in synsets1:
        for syn2 in synsets2:
            simi_score = syn1.path_similarity(syn2)
            if simi_score is not None:
                arr_simi_score.append(simi_score)
                best_score = max(arr_simi_score)
        score += best_score
        count += 1
    
    # Average the values
    score /= count
    return score
```

```python
sencentes_one = random.choice(review_data_list)
sencentes_two = random.choice(review_data_list)
#print("Sentence_one: [ {} ], \n\nSentence_two: [ {} ],\n\nSimilarity score:[ {} ] ".format(text_one, text_two, sentence_similarity(text_one, text_two)))
print("\n___\n\n\n-> Similarity score:[ {} ] ".format(sentence_similarity(sencentes_one, sencentes_two)))
```
